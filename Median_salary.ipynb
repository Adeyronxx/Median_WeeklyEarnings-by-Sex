```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('Resources/gender_data.csv')
```

```python
df.head()
```

```python
# Check Nulls
df.isna().sum().sum()
```

```python
# Check for Duplicates
df.duplicated().sum()
```

```python
# Check datatypes
df.info()
```

```python
# Rename Columns

df.rename(columns= {'Number of workers (in thousands) - Total': 'TotalNumberofWorkers_InThousands',
                    'Number of workers (in thousands) - Men': 'NumberofWorkers_InThousands_Men',
                    'Number of workers (in thousands) - Women': 'NumberofWorkers_InThousands_Women',
                   'Median weekly earnings (in current dollars) - Total': 'MedianWeeklyEarningsDollars_Total',
                   'Median weekly earnings (in current dollars) - Men': 'MedianWeeklyEarningsDollars_Men',
                   'Median weekly earnings (in current dollars) - Women': 'MedianWeeklyEarningsDollars_Women',
                   'Median weekly earnings (in constant dollars) - Total': 'MedianWeeklyEarningsCnstDollars_Total',
                    'Median weekly earnings (in constant dollars) - Men': 'MedianWeeklyEarningsCnstDollars_Men',
                    'Median weekly earnings (in constant dollars) - Women': 'MedianWeeklyEarningsCnstDollars_Women'
                   },inplace=True)
df.head()
```

Replace Strings and update data types

```python
df['TotalNumberofWorkers_InThousands']= df['TotalNumberofWorkers_InThousands'].map(lambda x: x.replace(',',''))
df['TotalNumberofWorkers_InThousands']=df['TotalNumberofWorkers_InThousands'].astype('int')
```

```python
df['NumberofWorkers_InThousands_Men'] = df['NumberofWorkers_InThousands_Men'].map(lambda x: x.replace(',',''))
df['NumberofWorkers_InThousands_Men'] = df['NumberofWorkers_InThousands_Men'].astype('int')
```

```python
df['NumberofWorkers_InThousands_Women'] = df['NumberofWorkers_InThousands_Women'].map(lambda x: x.replace(',',''))
df['NumberofWorkers_InThousands_Women'] = df['NumberofWorkers_InThousands_Women'].astype('int')
```

```python
df[['MedianWeeklyEarningsCnstDollars_Total','MedianWeeklyEarningsCnstDollars_Men','MedianWeeklyEarningsCnstDollars_Women']].describe()
```

```python
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
sns.lineplot(data=df,x='Year',y='NumberofWorkers_InThousands_Women')
plt.show()
```

```python
sns.lineplot(data=df,x='Year',y='MedianWeeklyEarningsCnstDollars_Men')
plt.show()
```

```python
sns.lineplot(data=df,x='Year',y='MedianWeeklyEarningsCnstDollars_Women')
plt.show()
```

```python
new_df = df[['MedianWeeklyEarningsCnstDollars_Men','MedianWeeklyEarningsCnstDollars_Women','NumberofWorkers_InThousands_Men','NumberofWorkers_InThousands_Women','MedianWeeklyEarningsCnstDollars_Total']].groupby(df['Year']).mean()
new_df=new_df.reset_index()
```

```python
new_df
```

```python
workers_df = new_df.melt(id_vars='Year',
                  value_vars=['NumberofWorkers_InThousands_Women','NumberofWorkers_InThousands_Men'],
                  var_name= 'WorkersbySex',
                  value_name= 'Population')
```

```python
workers_df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=workers_df, x='Year',y='Population',hue='WorkersbySex')
plt.xlabel('Year')
plt.ylabel('Population in Thousands')
plt.show()
```

```python
earnings_df = new_df.melt(id_vars='Year',
                  value_vars=['MedianWeeklyEarningsCnstDollars_Women','MedianWeeklyEarningsCnstDollars_Men'],
                  var_name= 'EarningsbySex',
                  value_name= 'Earnings')
```

```python
earnings_df.to_csv('earningsbysex.csv')
```

```python
earnings_df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=earnings_df, x='Year',y='Earnings',hue='EarningsbySex')

plt.xlabel('Year')
plt.ylabel('Earnings in Dollars')

plt.show()
```
